```python
import torch
import pandas as pd
import numpy as np
import csv
import json
import time
import random
```

```python
train = []
with open('data/train_trad.json') as f:
    for line in f:
        train.append(json.loads(line))

with open('data/generatal_tweet.json') as f:
  for line in f:
        train.append(json.loads(line))

print('we have total', len(train), 'samples')
```

```python
test = []
with open('data/train_new.json') as f:
    for line in f:
        test.append(json.loads(line))

print('we have total', len(test), 'test samples')
```

```python
test[1]
```

## Seperate data for training samples

```python
train_hate = []
train_no_hate = []

for d in train:
    if d['label'] == 1:
        train_hate.append(d)
    else:
        train_no_hate.append(d)
print('we have total', len(train_hate), 'hateful samples in training set')
print('we have total', len(train_no_hate), 'not hateful samples in training set')
```

### hate training samples

```python
# sentiment
train_hate_sent = []
for d in train_hate:
    if d['sentiment'] == -1:
        train_hate_sent.append(d)
print(len(train_hate_sent))    
```

```python
# target
train_hate_tar = []
for d in train_hate:
    if d['target'] == 1:
        train_hate_tar.append(d)
print(len(train_hate_tar))  
```

```python
# othering
train_hate_otr = []
for d in train_hate:
    if d['othering'] == 1:
        train_hate_otr.append(d)
print(len(train_hate_otr))  
```

```python
# derogatory term
train_hate_der = []
for d in train_hate:
    if d['derogatory'] == 1:
        train_hate_der.append(d)
print(len(train_hate_der))  
```

```python
# threatening
train_hate_thr = []
for d in train_hate:
    if d['threat'] == 1:
        train_hate_thr.append(d)
print(len(train_hate_thr))  
```

```python
# if contains more than 3 of them
train_hate_3 = []
for d in train_hate:
    count = 0
    if d['sentiment'] == -1:
        count += 1
    if d['target'] == 1:
        count += 1
    if d['derogatory'] == 1:
        count += 1
    if d['othering'] == 1:
        count += 1
    if d['threat'] == 1:
        count += 1

    if count >= 3:
        train_hate_3.append(d)
        
print(len(train_hate_3))  
```

### non-hate training samples

```python
# if contains more than 3 of them
# remove the sample from the dataset
train_no_hate_3 = []
for d in train_no_hate:
    count = 0
    if d['sentiment'] == -1:
        count += 1
    if d['target'] == 1:
        count += 1
    if d['derogatory'] == 1:
        count += 1
    if d['othering'] == 1:
        count += 1
    if d['threat'] == 1:
        count += 1

    if count >= 3:
        train_no_hate_3.append(d)
        
print(len(train_no_hate_3))  
```

```python
# print(len(train_no_hate))
for d in train_no_hate:
    if d in train_no_hate_3:
        train_no_hate.remove(d)
print(len(train_no_hate))
```

```python
with open('train_hate.json', 'w') as fout:
    json.dump(train_hate, fout)
```

```python
with open('train_nohate.json', 'w') as fout:
    json.dump(train_no_hate, fout)
```
